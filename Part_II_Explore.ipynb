```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe
df = pd.read_csv('prosper_clean.csv')
```

```python
# visualization codes
# function for first visual
def viz1():
    plt.figure(figsize=(15,6))
    
    # base color palette
    base_color = sb.color_palette()[2]
    sb.countplot(data=df, y='LoanStatus', hue='IsBorrowerHomeowner', color=base_color)
    plt.title('LoanStatus vs IsBorrowerHomeowner')
    plt.xticks(rotation=45)
    
# function for second visualization
def viz2():
    plt.figure(figsize=(15,6))
    plt.subplot(1,2,2)

    plt.subplot(1,2,1)
    sb.lineplot(data=df, x="LoanOriginationYear", y="BorrowerAPR", hue='IsBorrowerHomeowner')
    plt.title('BorrowerAPR vs LoanOriginationYear')

    plt.subplot(1,2,2)
    sb.lineplot(data=df, x="LoanOriginationYear", y="LoanOriginalAmount", hue='IsBorrowerHomeowner')
    plt.title('LoanOriginalAmount vs LoanOriginationYear')

# function for third visualization
def viz3():
    plt.figure(figsize=(15,10))
    plt.subplot(2,1,1)

    plt.subplot(2,1,1)
    sb.lineplot(data=df, x="LoanOriginationYear", y="BorrowerAPR", style='LoanStatus')
    plt.title('BorrowerAPR vs LoanOriginationYear vs LoanStatus')

    plt.subplot(2,1,2)
    sb.lineplot(data=df, x="LoanOriginationYear", y="LoanOriginalAmount", style='LoanStatus')
    plt.title('LoanOriginalAmount vs LoanOriginationYear vs LoanStatus')
```

## Visualization 1

## (Visualization 1)
- With the exception of `Current` loans, all other `Loan Status` are dominated by those who do not own homes, this as we would see in the next visuals is largely because Home owners can afford to take more loans, while those who do not own homes have to worry (possibly) about rent, and other financial commitments.

```python
viz1()
```

## (Visuaization 2)
- Here, we see an interesting relationship between `Homeownership`, `Loan original amount` and `Borrower Apr`.
- Home owners have significantly lower `Borrower APR` attatched to their loans while those who do not own homes have higher `Borrower APR`.
- Also, those who own homes take on larger loan amounts than those who do not own any homes, This could be because those who own homes have less to worry about rent and can channel those funds to more productive ventures on which the loans taken complement profitability.

```python
viz2()
```

## (Visualization 3)
- Here,  we also see an interesting playout for `Completed` and `currrent` loans.
- For the pre-2011 years, wee the `BorrowerAPR` steadily increasing and peaks in the year 2011 and then declines for all `LoanStatus`.
- However, during this same period, we see an oscillating pattern for other `LoanStatus` with the exception of `Current` loans which experiences a sharp decline and a gradually recovers with a decrease in `Borrower APR`.

```python
viz3()
```
